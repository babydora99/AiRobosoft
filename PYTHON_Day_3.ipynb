```python
#check whether a number is prime or not
def is_prime(num):
    for n in range(2,num):
        if num%n==0:
            print("Not prime")
            break
    else:
        print("Prime")
```

```python
is_prime(11)
```

```python
import math
def is_prime(num):
    if num%2==0 and num>2:
        return False
    for i in range(3,int(math.sqrt(num))+1,2):
        if num%i==0:
            return False
    return True
```

```python
is_prime(25)
```

## Iterators and Generators


```python
#Generator function for the cube of numbers 
def genCube(n):
    for num in range(n):
        yield num**3
```

```python
for x in genCube(10):
    print(x)
```

```python
def fibonacci(num):
    a=1
    b=1
    for  i in range(num):
        yield a
        a,b=b,a+b
```

```python
for x in fibonacci(5):
    print(x)
```

```python
#Normal code
def fibo(num):
    a=1
    b=1
    output=[]
    for i in range(num):
        output.append(a)
        a,b=b,a+b
    return output
```

```python
fibo(10)
```

### map()
a map is a function that takes in two argument
1.A function
2.A sequence iterable

In the form:map(function,sequence)
The first argument is the name of the function and the second is a sequence(e.g a list).map() applies the function to all the elements of the sequence.

```python
#convert fahreheit to celsius
def farh(T):
    return ((float(9)/5)*T+32)
def celc(T):
    return (float(5)/9)*(T-32)

temp=[0,22.5,40,100]
```

```python
#use map() function
f_temps=list(map(farh,temp))
f_temps
```

```python
#convert it back
list(map(celc,f_temps))
```

### Lambda function
it is a nameless function

```python
list(map(lambda x:(5.0/9)*(x-32),f_temps))
```

### filter() function
it filters out all the elements of an iterable.Syntax:filter(function,sequence),for which the function returns 'True'.



```python
def even_lst(num):
    if num%2==0:
        return True
```

```python
lst=range(15)
my_lst=list(lst)
print(my_lst)
```

```python
list(filter(even_lst,lst))
```

```python
list(filter(lambda x:x%2==0,my_lst))
```
