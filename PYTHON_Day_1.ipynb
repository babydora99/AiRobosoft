## For loop :
#### it is used to itearte over a sequence (list/tuple/String) .iterating over a sequence is called Traversal.

```python
#Program to find the sum of variables stored in a list 

#list 
numbers=[1,2,3,4,5]

#variable to store the sum
sum=0

#Iteration of the list
for val in numbers:
    sum=sum+val

#print the output
print("The sum is",sum)
```

## for loop with else : 

```python
for val in numbers:
    print(val)
else:
    print("No items left")
```

## while loop
#### while loop is used to iterate over a block of code as long as the text expression(condition) is TRUE

```python
#print the sum of n numbers

#take a user input
n=int(input("ENter the number of n : "))

#initialise the sum and counter
sum=0
i=1

#while loop:
while(i<=n):
    sum=sum+i
    i=i+1

#print the sum
print("The sum is:",sum)
```

## range() function

we can generate a sequence of numbers using range() function  , it defines range(start,stop,step size) .if step size is not defined then it defaults to 1

```python
print(range(10))
```

```python
print(list(range(10))) # print the numbers within the range
```

```python
print(list(range(2,8))) # print the range start from 2 and stop at 8
```

```python
print(list(range(2,29,3))) # print the range start at 2 and stop at 9 and the gap between them should be 3 (size between them is 3)
```

we can use the range() function in for loop to get the sequence of numbers .it can be  combined with the len() to iterate through a sequence using indexing. 

```python
#Program to iterate through a list using indexing 
genre=['apple','mango','grapes','jack']
for val in range(len(genre)):
    print("I love ",genre[val])
```

## break and continue statement

in python ,break nad continue statement is used to alter the normal loop


### break statement
break statement is used to terminate the loop containing it.

```python
for val in "VARSIKA":
    if val=="I":
        break
    print(val)

print("The End")
```

### continue statement 
continue state is used to skip the rest of the code inside the loop for the current iteration only . loops doesnot terminate but continue on next iteration

```python
for val in "VARSIKA":
    if val=="I":
        continue
    print(val)

print("The End")
```
