#### Methods

```python
class Person:
    def __init__(self,name,surname,year_Of_birth):
        self.name=name
        self.surname=surname
        self.year_Of_birth=year_Of_birth

    def age(self,current_year):
        return current_year-self.year_Of_birth

    def __str__(self):
        return '%s %s was born in %d. ' %(self.name,self.surname,self.year_Of_birth)

alec=Person('baby','dora',2004)
print(alec)
print(alec.age(2025))
```

Two methods are used i.e age and _ _str_ _.

it is called to represent the object as String,when it has to be printed.



### Inheritance

```python
class Student(Person):
    def __init__(self,student_id,*args,**kwargs):
        super(Student,self).__init__(*args,**kwargs)
        self.student_id=student_id

charlie=Student(1,"baby","dora",2019)
print(charlie)
print(type(charlie))
print(isinstance(charlie,Person))
print(isinstance(charlie,object))
    
```

Here charlie is the instance of the Person.
charlie is the instance of object ,bcz charlie is the instance of Student class which is main class .

Also Person is the base_class of Student.Student is the sub class of Person.

super(class,instance)

#### Overriding Methods
Inheritance allowed to add new method to the subclass but it is basically used to change the behaviour of the method used in subclass.

To override a method just define it again.

```python
class Student(Person):
    def __init__(self,student_id,*args,**kwargs):
        super(Student,self).__init__(*args,**kwargs)
        self._student_id=student_id

    def __str__(self):
        return super(Student,self).__str__()+"And has ID: %d " %self._student_id

charlie=Student(1,"baby","dora",2004)
print(charlie)
```

## StringIO
The StringIO module implements an in-memory filelike object. This object can then be used as input or output to most functions that would expect a standard file object.


```python
from io import StringIO
```

```python
#Arbitrary String
msg='This is just a normal string'
```

```python
#use StringIO method to set as file object
t=StringIO(msg)
```

Now we have an object *f* that we will be able to treat just like a file. For example:

```python
t.read()
```

```python
t.write(' Second line written to file like object')
```

```python
#Reset cursor just like you wound a file
t.seek(0)
```

```python
#read again
t.read()
```
